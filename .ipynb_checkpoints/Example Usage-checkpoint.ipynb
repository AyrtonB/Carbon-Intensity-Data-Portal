### Advanced Usage

Often we want more than just information on todays emissions

```python
df = wrapper.query_API('2020-01-01', '2020-01-02')

df.head()
```

<br>

The Carbon Intensity API doesn't allow requests for data that exceeds 2-weeks, fortunately the Python wrapper handles this for us automatically

```python
df = wrapper.query_API('2020-01-01', '2020-01-31')

df.head()
```

<br>

We can also request data at a higher spatial resolution, in this case for the DNO regions

N.b. at time of writing the Carbon Intensity API only returns forecast emissions at a regional level

```python
df = wrapper.query_API(level='regional')

df.head()
```

<br>

We can also get generation data disaggregated by fuel-type

```python
df = wrapper.query_API(data_stream='generation')

df.head()
```

<br>

Putting these together you can make very specific queries, here we'll ask for regional generation in January 2020

```python
df = wrapper.query_API('2020-01-01',
                       '2020-01-31',
                       level='regional', 
                       data_stream='generation')

df.head()
```
